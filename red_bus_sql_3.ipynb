```python
# csv to dataframe
df_buses_1=pd.read_csv("Kerala Bus details.csv")
df_buses_2=pd.read_csv("Andhra Bus details.csv")
df_buses_3=pd.read_csv("Telangana Bus details.csv")
df_buses_4=pd.read_csv("Rajasthan Bus details.csv")
df_buses_5=pd.read_csv("Bihar Bus details.csv")
df_buses_6=pd.read_csv("South Bengal Bus details.csv")
df_buses_7=pd.read_csv("Himachal_pradesh Bus details.csv")
df_buses_8=pd.read_csv("Chandigarh Bus details.csv")
df_buses_9=pd.read_csv("Uttar pradesh Bus details.csv")
df_buses_10=pd.read_csv("Punjab  Bus details.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

```python
Final_df.info()
```

```python
Final_df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
Final_df
```

```python
#convert prices to numeric
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
```

```python
Final_df["Price"].fillna(0,inplace=True)
```

```python
Final_df["Price"]
```

```python
#convert Ratings to numeric
Final_df["Ratings"]=Final_df["Ratings"].str.replace("New","")
Final_df["Ratings"]=Final_df["Ratings"].str.strip()
Final_df["Ratings"]=Final_df["Ratings"].str.split().str[0]
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
Final_df["Ratings"].fillna(0,inplace=True)
```

```python
Final_df["Ratings"]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
def insert_data_into_postgresql(data):
    try:
        # Connect to PostgreSQL
        connection = psycopg2.connect(
            user="postgres",  # Replace with your PostgreSQL username
            password="harish26",  # Replace with your PostgreSQL password
            port="5433",  # Replace with your PostgreSQL port
            database="postgres"  # Replace with your PostgreSQL database name
        )
        cursor = connection.cursor()        
        
        # Insert data into PostgreSQL table
        for index, row in data.iterrows():
            insert_query = """
            INSERT INTO bus_details (Bus_name, Bus_type, Start_time, End_time, Total_duration, Price, Seats_available,Ratings,Route_links,Route_name)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            """
            record_to_insert = (
                str(row['Bus_name']) if row['Bus_name'] else None,  # Handle None properly
                str(row['Bus_type']) if row['Bus_type'] else None,
                str(row['Start_time']) if row['Start_time'] else None,
                str(row['End_time']) if row['End_time'] else None,
                str(row['Total_duration']) if row['Total_duration'] else None,
                str(row['Price']) if row['Price'] else None,
                str(row['Seats_Available']) if row['Seats_Available'] else None,
                str(row['Ratings']) if row['Ratings'] is not None else None,  # Ensure Ratings is a float
                str(row['Route_link']) if row['Route_link'] else None,
                str(row['Route_name']) if row['Route_name'] else None)
            cursor.execute(insert_query, record_to_insert)
        
        # Commit the transaction
        connection.commit()
        print("Data successfully inserted into PostgreSQL")
    
    except (Exception, psycopg2.Error) as error:
        print(f"Failed to insert data into PostgreSQL: {error}")
    
    finally:
        # Close connection
        if connection:
            cursor.close()
            connection.close()

# Call the function to insert data into PostgreSQL
insert_data_into_postgresql(bus_df)
```
